## Q2: Implement Gradient Descent

Use the function $f(x) = x^2 -4x + 3$ as the cost function.
Perform the following steps for implementing gradient descent on the cost function.
1. **Define the Cost Function:**
   - Create a function `cost_function(x)` that returns the value of a quadratic function $f(x) = x^2 - 4x + 3$, which is the objective function we want to minimize.

2. **Define the Gradient:**
   - Write a function `derivative(x)` to compute the derivative of the cost function. The derivative is $f'(x) = 2x - 4$, which represents the slope at any point \( x \).

3. **Set Up Gradient Descent:**
   - Define a function `gradient_descent(learning_rate, initial_x, iterations)` to implement the gradient descent algorithm.
   - Initialize `x` with the starting value `initial_x`.

4. **Gradient Descent Loop:**
   - Use a `for` loop that runs for `iterations` number of times.
   - Inside the loop, calculate the gradient using the `derivative(x)` function.
   - Update the value of `x` by subtracting `learning_rate * gradient` to move \( x \) in the direction of minimizing the cost.
   - Calculate and print the current value of the cost function using `cost_function(x)` and display the iteration number.

5. **Return the Optimal Value:**
   - After the loop finishes, return the final value of \( x \), which is the estimated optimal point.

6. **Set Parameters:**
   - Set the `learning_rate`, `initial_x`, and `iterations` as needed for the gradient descent function.

7. **Run Gradient Descent:**
   - Call `gradient_descent()` with the appropriate parameters and print the final optimal value of \( x \) and the corresponding cost.

8. **Verify Result:**
   - Print the final `cost_function(optimal_x)` to confirm the cost at the optimal point.

```python
# Define the cost function
def cost_function(x):
    return x**2 - 4*x + 3

# Define the derivative of the cost function
def derivative(x):
    return 2*x - 4

# Define the gradient descent function to minimize the cost function
def gradient_descent(learning_rate, initial_x, iteration):
    x = initial_x

    # Iterate for the given number of iterations
    for i in range(iterations):
        # Calculate the gradient of the cost function at current x
        gradient = derivative(x)

        # Update x using gradient descent
        x = x - learning_rate * gradient

        # Calculate the value of the cost function at the updated x
        cost = cost_function(x)

        # Print the current iteration, the value of x, and the value of the cost function
        print("Iteration:", i, " x =", x, " Cost =", cost)

    # Return the final value of x
    return x

# Parameters
learning_rate = 0.1  # Step size for gradient descent
initial_x = 0        # Initial value of x
iterations = 100      # Number of iterations

# Run gradient descent
optimal_x = gradient_descent(learning_rate, initial_x, iterations)

# Print the optimal value of x and the corresponding cost
print("Optimal x:", optimal_x)
print("Optimal cost:", cost_function(optimal_x))
```

### Interpretation:
- The gradient descent output confirms our observation in the plot that minima is at x=2.

## Q3: Implement gradient descent on a real-life data
Apply all 3 types of gradient descent on the breast cancer data available in `sklearn`.
- Batch Gradient Descent
- Mini Batch Gradient Descent
- Stochaistic Gradient Descent

### Steps for Implementing Gradient Descent with Logistic Regression:

1. **Load and Preprocess Dataset:**
   - Load the **Breast Cancer** dataset using `load_breast_cancer()`.
   - Standardize the features using `StandardScaler()` to scale the data for gradient descent.
   - Split the dataset into training and testing sets using `train_test_split()`.

2. **Add Bias Term:**
   - Add a column of ones to `X_train` and `X_test` to account for the bias term (intercept).

3. **Define Core Functions:**
   - **Sigmoid Function:** `sigmoid(z)` to compute the logistic function.
   - **Cost Function:** `compute_cost(X, y, theta)` to compute binary cross-entropy loss.
   - **Gradient Function:** `compute_gradient(X, y, theta)` to calculate the gradient.

4. **Implement Gradient Descent Methods:**
   - **Batch Gradient Descent:** `batch_gradient_descent()` updates weights using the entire dataset.
   - **Mini-Batch Gradient Descent:** `mini_batch_gradient_descent()` updates weights using small random batches.
   - **Stochastic Gradient Descent:** `stochastic_gradient_descent()` updates weights using a single random sample per iteration.

5. **Set Hyperparameters:**
   - Initialize weights `theta_initial` to zeros.
   - Set learning rate `alpha = 0.1`, iterations `1000`, and mini-batch size `32`.

6. **Run Gradient Descent:**
   - Execute all three methods (Batch, Mini-Batch, and Stochastic Gradient Descent) and track cost history.

7. **Plot Results:**
   - Plot the cost history for all three methods on the same graph to visualize convergence behavior.

8. **Make Predictions:**
   - Use the `predict()` function to make predictions on the test set with each model.

9. **Evaluate Models:**
   - Print classification reports for all three methods (`Batch`, `Mini-Batch`, and `Stochastic Gradient Descent`) using `classification_report()` to compare accuracy.



```python
# Import the required libraries
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')


# Load the Breast Cancer dataset
cancer = load_breast_cancer()
X = cancer.data
y = (cancer.target != 0) * 1  # Binary classification (0 or 1)

# Standardize the dataset (important for gradient descent)
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Add intercept term (bias) to X
X_train_b = np.c_[np.ones((X_train.shape[0], 1)), X_train]  # Add a column of ones for intercept
X_test_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]     # Same for test set

# Sigmoid function for logistic regression
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Logistic regression cost function (binary cross-entropy)
def compute_cost(X, y, theta):
    m = len(y)
    h = sigmoid(X.dot(theta))
    cost = -(1/m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))
    return cost

# Gradient of the cost function
def compute_gradient(X, y, theta):
    m = len(y)
    h = sigmoid(X.dot(theta))
    gradient = (1/m) * X.T.dot(h - y)
    return gradient

# Batch Gradient Descent
def batch_gradient_descent(X, y, theta, alpha, iterations):
    cost_history = []
    for i in range(iterations):
        gradient = compute_gradient(X, y, theta)
        theta -= alpha * gradient
        cost_history.append(compute_cost(X, y, theta))
    return theta, cost_history

# Mini-Batch Gradient Descent
def mini_batch_gradient_descent(X, y, theta, alpha, iterations, batch_size):
    m = len(y)
    cost_history = []
    for i in range(iterations):
        indices = np.random.permutation(m)
        X_shuffled = X[indices]
        y_shuffled = y[indices]
        for j in range(0, m, batch_size):
            X_mini_batch = X_shuffled[j:j+batch_size]
            y_mini_batch = y_shuffled[j:j+batch_size]
            gradient = compute_gradient(X_mini_batch, y_mini_batch, theta)
            theta -= alpha * gradient
        cost_history.append(compute_cost(X, y, theta))
    return theta, cost_history

# Stochastic Gradient Descent
def stochastic_gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    cost_history = []
    for i in range(iterations):
        for j in range(m):
            rand_index = np.random.randint(0, m)
            X_i = X[rand_index, :].reshape(1, X.shape[1])
            y_i = y[rand_index].reshape(1)
            gradient = compute_gradient(X_i, y_i, theta)
            theta -= alpha * gradient
        cost_history.append(compute_cost(X, y, theta))
    return theta, cost_history

# Prepare initial theta (weights)
n = X_train_b.shape[1]
theta_initial = np.zeros(n)

# Hyperparameters
alpha = 0.1
iterations = 1000
batch_size = 32  # For mini-batch

# ----------- Batch Gradient Descent -----------
theta_bgd, cost_history_bgd = batch_gradient_descent(X_train_b, y_train, theta_initial.copy(), alpha, iterations)

# ----------- Mini-Batch Gradient Descent -----------
theta_mbgd, cost_history_mbgd = mini_batch_gradient_descent(X_train_b, y_train, theta_initial.copy(), alpha, iterations, batch_size)

# ----------- Stochastic Gradient Descent -----------
theta_sgd, cost_history_sgd = stochastic_gradient_descent(X_train_b, y_train, theta_initial.copy(), alpha, iterations)

# Plot the cost history for each method
plt.figure(figsize=(6, 6))
plt.plot(range(iterations), cost_history_bgd, label='Batch Gradient Descent', color='r')
plt.plot(range(iterations), cost_history_mbgd, label='Mini-Batch Gradient Descent', color='g')
plt.plot(range(iterations), cost_history_sgd, label='Stochastic Gradient Descent', color='b')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost History for Different Gradient Descent Methods')
plt.legend()
plt.show()

# Predict function for logistic regression
def predict(X, theta):
    return np.round(sigmoid(X.dot(theta)))

# Evaluate all three models
y_pred_bgd = predict(X_test_b, theta_bgd)
y_pred_mbgd = predict(X_test_b, theta_mbgd)
y_pred_sgd = predict(X_test_b, theta_sgd)

# Print classification reports
print("Classification Report for Batch Gradient Descent:\n")
print(classification_report(y_test, y_pred_bgd))

print("\nClassification Report for Mini-Batch Gradient Descent:\n")
print(classification_report(y_test, y_pred_mbgd))

print("\nClassification Report for Stochastic Gradient Descent:\n")
print(classification_report(y_test, y_pred_sgd))
```

### Interpretation:

1. **Batch Gradient Descent (Red Line)**:
   - Starts with a high initial cost.
   - Converges steadily but more slowly compared to other methods.
   - Requires more iterations to reach a lower cost due to updating weights after processing the entire dataset.

2. **Mini-Batch Gradient Descent (Green Line)**:
   - Converges faster than Batch Gradient Descent.
   - Offers a balance between stability and speed, using smaller mini-batches.
   - Cost decreases smoothly without much fluctuation.

3. **Stochastic Gradient Descent (Blue Line)**:
   - Converges the fastest but with high fluctuations due to frequent updates after processing individual samples.
   - The cost fluctuates around the minimum, indicating the noisy nature of updates with small learning rate (0.1).
   - Stabilizes quickly but shows more variance in cost.

4. **Overall Observation**:
   - Stochastic Gradient Descent achieves faster convergence but with instability.
   - Mini-Batch Gradient Descent is smoother and converges relatively quickly.
   - Batch Gradient Descent is slower but stable throughout.
