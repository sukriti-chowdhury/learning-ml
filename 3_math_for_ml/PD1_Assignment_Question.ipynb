You are analyzing the number of customer service calls received at a call center, which follows a Poisson distribution with an average of 5 calls per hour. Calculate the probability of receiving exactly 3 calls in the next hour.

**Instructions:**
Write Python code to compute the probability of a specific number of events (calls) in a Poisson distribution. Plot the probability mass function (PMF) and cumulative distribution function (CDF) for visualization.

```python
#Import the required modules
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Parameters
lambda_ = 5  # Average number of calls per hour
k = 3  # Number of calls we are interested in

# Probability of exactly 3 calls in the next hour
prob_3_calls = poisson.pmf(k, lambda_)

# Plot the PMF
# Hint: Use poisson.pmf()


# Plot the CDF
# Hint: Use poisson.cdf()


# Print resultant probability
print(f"Probability of exactly 3 calls in the next hour: {prob_3_calls:.4f}")
```
