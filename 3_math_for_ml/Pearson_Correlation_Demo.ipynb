# Objective

- To understand how to calculate Pearson correlation coefficient using 3 different libraries

### Import Libraries

```python
import numpy as np
import pandas as pd
import scipy.stats
import seaborn as sns
```


### Load car crash dataset

```python
df = sns.load_dataset("car_crashes")
```

```python
df.head()
```

#### Let's look at the graph of how does alcohol influence affect driving speed

```python
sns.scatterplot(df, x='speeding', y='alcohol')
```

## 1. Finding correlation using Pandas

#### Corr function in pandas gives correlatiob matrix for all feature pairs

```python
# Drop the non-numeric columns for correlation analysis
df_numeric = df.select_dtypes(include=[np.number])
df_numeric
```

```python
df_numeric.corr()
```

- It supports other correlation coefficient as well

## 2. Finding correlation using Scipy library

```python
scipy.stats.pearsonr(df_numeric['speeding'], df_numeric['alcohol'])
```

### Let's pickup another set of feature pairs

```python
sns.scatterplot(df_numeric, x='ins_premium', y='alcohol')
```

```python
scipy.stats.pearsonr(df_numeric['ins_premium'], df_numeric['alcohol'])[0]
```

##### PCC is symmetrics, look at the below swap

```python
scipy.stats.pearsonr(df_numeric['alcohol'], df_numeric['ins_premium'])[0]
```

### PCC between number of accidents and alcohol influence

```python
scipy.stats.pearsonr(df_numeric['alcohol'], df_numeric['total'])[0]
```

```python
sns.scatterplot(df_numeric, x='total', y='alcohol')
```

## 3. Using Numpy to find PCC

```python
print("PCC using numpy: ", np.corrcoef(df_numeric['alcohol'], df_numeric['total']))
```
